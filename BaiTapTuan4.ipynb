Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

try:
    line = input().split()
    if len(line) == 2:
        a = int(line[0])
        b = int(line[1])
        sum = 0
    for i in range(a, b+1):
        if is_prime(i):
            sum += i
    print(sum)
except EOFError:
    pass
```

[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).

```python
n_str = input()
if n_str:
    n = int(n_str)
    results = []
    i = 0
    
    # Tìm các số chính phương <= n
    while True:
        square = i * i
        if square > n:
            break
        
        # Kiểm tra các chữ số đôi một khác nhau
        s = str(square)
        is_distinct = True
        # Duyệt qua từng chữ số để kiểm tra trùng lặp
        for char in s:
            if s.count(char) > 1:
                is_distinct = False
                break
        
        if is_distinct:
            results.append(str(square))
        i += 1

    # In kết quả theo định dạng yêu cầu
    if len(results) == 0:
        print("(no number)")
    else:
        print(" ".join(results))

```

W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
n = int(input())

max_length = 0
best_x = 0

for x in range(1, n + 1):
    current_n = x
    length = 1
    while current_n != 1:
        if current_n % 2 == 0:
            current_n = current_n // 2
        else:
            current_n = 3 * current_n + 1
        length += 1
    if length > max_length:
        max_length = length
        best_x = x

if n > 0:
    print(f"{best_x} {max_length}")
```

Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b

```python
a, b = map(int, input().split())

for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

Bài 10:
Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính n

```python
a = int(input())
tong = 0
for i in range(1, a):
    if a % i == 0:
        tong += i
if tong == a:
    print("True")
else:
    print("False")
```

Bài 14:
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
min = 0
max = 0
while True:
    n = int(input())
    if n != -1:
        if n>max:
            max = n
        if n<min:
            min = n
        continue
    break
print(max)
print(min)
```

Bài 15:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều cách nhất có thể, ko dùng mảng.

```python
n = int(input())
count = 0
while n != 0:
    n //= 10
    count += 1
print(count)
```

Bài 16:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
while n != 0:
    chu_so = n % 10
    n //= 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
print(chan, "chan", le, "le")
```

Bài 19:
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó.
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
f1, f2 = 1, 1
while f1 + f2 <= A:
    f_next = f1 + f2
    f1 = f2
    f2 = f_next
print(f2)
```

Bài 20:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
import math
n = int(input())
count = 0
while n != 0:
    chu_so = n % 10
    n //= 10
    if chu_so < 2:
        count += 0
    else:
        for i in range(2, math.isqrt(chu_so) + 1):
            if chu_so % i == 0:
                count += 0
        count += 1
print(count)
```
