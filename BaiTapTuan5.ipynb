Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý 
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and 
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong

trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc 
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết 
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác 
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ 
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách 
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh 
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự 
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có 
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 
'z', 'b' thành 'b' và 'c' thành 'x'. 
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai 
từ đầu vào có cùng độ dài. 


```python
def is_isomorphic(a, b):
    d = {}
    s = set()
    for x, y in zip(a, b):
        if x in d:
            if d[x] != y:
                return False
        else:
            if y in s:
                return False
            d[x] = y
            s.add(y)
    return True


print(is_isomorphic("abca", "zbxz"))
print(is_isomorphic("abc", "dee"))
print(is_isomorphic("egg", "add"))
print(is_isomorphic("paper", "title"))
print(is_isomorphic("foo", "bar"))
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_palindrome(x):
    if x < 0:
        return False
    
    original = x
    reversed_num = 0
    
    while x > 0:
        reversed_num = reversed_num * 10 + x % 10
        x //= 10
        
    return original == reversed_num

# Thử nghiệm
print(is_palindrome(121))   # True
print(is_palindrome(12321)) # True
print(is_palindrome(123))   # False
print(is_palindrome(-121))  # False
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_magic_number(n):
    while n >= 10:
        tong = 0
        while n > 0:
            tong += n % 10
            n //= 10
        n = tong
    return n == 1

# Thử nghiệm
print(is_magic_number(163)) # True (1+6+3=10 -> 1+0=1)
print(is_magic_number(199)) # True (1+9+9=19 -> 1+9=10 -> 1+0=1)
print(is_magic_number(225)) # False (2+2+5=9)
```

W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)


```python
def point_in_quadrilateral(quad, p):
    def cross_product(a, b, c):
        return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])

    n = len(quad)
    signs = []

    for i in range(n):
      
        p1 = quad[i]
        p2 = quad[(i + 1) % n]
        
        cp = cross_product(p1, p2, p)
        
        if cp != 0:
            signs.append(cp > 0)

    # Nếu tất cả các tích chéo có cùng dấu (đều > 0 hoặc đều < 0), điểm nằm bên trong
    # Nếu có cp == 0, các dấu còn lại vẫn phải đồng nhất
    return len(set(signs)) <= 1

# Thử nghiệm
quad = [(0, 0), (4, 0), (4, 3), (0, 3)]
p1 = (2, 1)  # Nằm trong
p2 = (5, 5)  # Nằm ngoài
p3 = (4, 1)  # Nằm trên biên

print(point_in_quadrilateral(quad, p1)) # True
print(point_in_quadrilateral(quad, p2)) # False
print(point_in_quadrilateral(quad, p3)) # True
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math

def calculate_angle(x1, y1, x2, y2, x3, y3):
    ax, ay = x1 - x2, y1 - y2
    cx, cy = x3 - x2, y3 - y2
    
    dot = ax * cx + ay * cy
    mag_a = math.sqrt(ax**2 + ay**2)
    mag_c = math.sqrt(cx**2 + cy**2)
    
    if mag_a * mag_c == 0:
        return 0.0
        
    cos_theta = max(-1.0, min(1.0, dot / (mag_a * mag_c)))
    return math.degrees(math.acos(cos_theta))

# Cách dùng:
print(calculate_angle(0,1, 0,0, 1,0))
```

```python
import math

def nCr(n, k):
    if k < 0 or k > n:
        return 0
    return math.comb(n, k)

print(nCr(5, 2))
```

W5A31 Viết hàm tính tổng bình phương các số từ 1 đến n.


```python
def sum_squares(n):
    return sum(i**2 for i in range(1, n + 1))

print(sum_squares(5))
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.


```python
def tinh_tien_taxi(loai_xe, gio_di, so_km, phut_cho):
    if loai_xe == 4:
        mo_cua = 12000
        gia_km = 15000
    else:
        mo_cua = 15000
        gia_km = 18000

    phi_cho = phut_cho * 500
    tong_tien = mo_cua + (so_km * gia_km) + phi_cho

    if gio_di >= 22 or gio_di < 5:
        tong_tien *= 1.1

    if so_km > 30:
        tong_tien *= 0.9

    return round(tong_tien)

print(tinh_tien_taxi(4, 23, 35, 10))
```

W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    
    tong = 0
    for x in range(1, n + 1):
        s = str(x)
        if len(s) == len(set(s)):
            tong += x
            
    return tong

print(sum_1_to_n(20))
```
