[Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
cách nhau một dấu cách


```python
n = float(input())
nguyen = int(n)
du = n - nguyen
if n < 0 and du != 0:
    xuong = nguyen - 1
else:
    xuong = nguyen
if n > 0 and du != 0:
    len_en = nguyen + 1
else:
    len_en = nguyen
temp = n + 0.5
temp_nguyen = int(temp)
if temp < 0 and (temp - temp_nguyen) != 0:
    gan_nhat = temp_nguyen - 1
else:
    gan_nhat = temp_nguyen

print(f"{len_en} {xuong} {gan_nhat}")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math

x0 = float(input("Nhập hoành độ tâm I (x0): "))
y0 = float(input("Nhập tung độ tâm I (y0): "))
r = float(input("Nhập bán kính R: "))

xa = float(input("Nhập hoành độ điểm A (x): "))
ya = float(input("Nhập tung độ điểm A (y): "))

khoang_cach = math.sqrt((xa - x0)**2 + (ya - y0)**2)

if khoang_cach == r:
    print(f"Điểm A({xa}, {ya}) nằm TRÊN đường tròn.")
elif khoang_cach < r:
    print(f"Điểm A({xa}, {ya}) nằm TRONG đường tròn.")
else:
    print(f"Điểm A({xa}, {ya}) nằm NGOÀI đường tròn.")
```
