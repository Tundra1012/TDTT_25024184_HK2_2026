## W2A3 [ModAndDiv]
Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. 
In ra tổng, hiệu, tích, phần nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập phân) theo định dạng đúng của output

+: cộng, -: trừ, *: nhân, /:chia thực (5/2 = 2,500000), //: chia lấy (5//2 = 2) 
% (mod): chia dư (5%2 == 1), : mũ (2**3 = 8)


```python
a, b = map(int, input().split())
print(f"{a} + {b} = {a+b}")
print(f"{a} - {b} = {a-b}")
print(f"{a} * {b} = {a*b}")
print(f"{a} / {b} = {a/b:.2f}")
```

## W2A4 [AverageOfSix]
Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng
6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
• Các điểm a1, b1, c1 được nhân hệ số 1,
• Các điểm a2, b2 được nhân hệ số 2,
• Điểm a3 được nhân hệ số 3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6
đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được
phân tách nhau bởi duy nhất 1 dấu cách.
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.

```python
a1, b1, c1, a2, b2, a3 = map(int, input().split())
TB = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10
print(f"{TB:.1f}")
```

W2A16 [SeasonsInYear]
Trong bài tập này, bạn sẽ sử dụng hàm print để in nhiều dòng ra màn hình. Có nhiều cách để
làm việc này, cách thứ nhất là sử dụng lệnh print nhiều lần để in nhiều dòng khác nhau. Ví
dụ, để in ra tên 4 mùa bằng tiếng Anh trên 4 dòng, bạn dùng 4 lệnh như sau:

print("Summer")
print("Autumn")
print("Winter")
Trong cách thứ hai, bạn sử dụng duy nhất một lệnh print và đưa liên tiếp các dòng chữ trên
ra như sau:

Lưu ý rằng đây là một lệnh duy nhất, khi gặp dấu ‘\n’ là một kí tự xuống dòng được in ra.
Ngoài ra, bạn hãy tự tìm hiểu xem có còn cách nào khác không.
Để luyện tập, hãy viết vào ô dưới đây chương trình in ra trên màn hình tên các mùa trong
năm bằng tiếng Anh. Mỗi mùa in trên một dòng, bắt đầu từ "Spring".

```python
print("Spring")
print("Summer")
print("Autumn")
print("Winter")
print("Spring\nSummer\nAutumn\nWinter")
print("""Spring
Summer
Autumn
Winter""")
```
