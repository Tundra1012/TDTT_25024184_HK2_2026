Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
a = float(input())
b = float(input())
c = float(input())

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(round(s, 1))
else:
    print("Khong phai 3 canh cua tam giac")
```

Bài 8:
Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này, tính tiền điện tháng này cho hộ biết:

Từ 0 – 50kWh: 1.984 đồng/kWh
Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng

```python
ten_chu_ho = input("Ten chu ho: ")
chi_so_cu = int(input("Chi so thang truoc: "))
chi_so_moi = int(input("Chi so thang nay: "))

so_kwh = chi_so_moi - chi_so_cu
tien = 0

if so_kwh <= 50:
    tien = so_kwh * 1984
elif so_kwh <= 100:
    tien = 50 * 1984 + (so_kwh - 50) * 2050
elif so_kwh <= 200:
    tien = 50 * 1984 + 50 * 2050 + (so_kwh - 100) * 2380
elif so_kwh <= 300:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (so_kwh - 200) * 2998
elif so_kwh <= 400:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (so_kwh - 300) * 3350
else:
    tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (so_kwh - 400) * 3460

tong_tien = round(tien * 1.08)

print(f"Ho va ten: {ten_chu_ho}")
print(f"Tien phai tra la: {tong_tien}")
```
